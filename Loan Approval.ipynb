```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
X[' education'] = encoder.fit_transform(X[' education'])
X[' self_employed'] = encoder.fit_transform(X[' self_employed'])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()
log_model.fit(X_train,y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy : ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
log_model.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

r_model = RandomForestClassifier()
r_model.fit(X_train,y_train)
```

```python
y_pred_r = r_model.predict(X_test)
```

```python
print("Accuracy : ",accuracy_score(y_test,y_pred_r))
classification_report(y_test,y_pred)
```

```python
from sklearn.ensemble import VotingClassifier

ensemble_model = VotingClassifier(
    estimators=[('lr',log_model),('rf',r_model)],
    voting='soft'
)
```

```python
ensemble_model.fit(X_train,y_train)
```

```python
y_pred_ensemble = ensemble_model.predict(X_test)
```

```python
print("Accuracy : ",accuracy_score(y_pred_ensemble,y_test))
classification_report(y_pred_ensemble,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
g_model = GradientBoostingClassifier()
```

```python
ensemble_tree = VotingClassifier(
    estimators=[('rf',r_model),('gb',g_model)],
    voting = 'soft'
)
```

```python
ensemble_tree.fit(X_train,y_train)
```

```python
X_train_df = pd.DataFrame(X_train,columns=X.columns)
X_test_df = pd.DataFrame(X_test,columns=X.columns)
```

```python
def predict_fn(data):
    return ensemble_tree.predict_proba(data)
```

```python
# Counter Factuals
```

```python
print("SAi")
```

```python
import dice_ml
data_df = X_train_df.copy()
data_df['loan_status'] = y_train
```

```python
d = dice_ml.Data(
    dataframe=data_df,
    continuous_features=list(X_train_df.columns),
    outcome_name='loan_status'
)
```

```python
m = dice_ml.Model(
    model=ensemble_tree,
    backend="sklearn"
)
```

```python
dice_exp = dice_ml.Dice(d,m)
```

```python
query_instance = X_test_df.iloc[[0]]
```

```python
cf = dice_exp.generate_counterfactuals(
    query_instance,
    total_CFs = 8,
    desired_class='opposite'
)
```

```python
cf.visualize_as_dataframe()
```

```python
cf.visualize_as_list()
```
